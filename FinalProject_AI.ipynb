# Preprocessing Data

```python
# 1. Import File

from google.colab import drive
drive.mount('/content/drive')

# Path file
train_path = '/content/drive/My Drive/KELAS AI/df_train.csv'
test_path = '/content/drive/My Drive/KELAS AI/df_test.csv'

# Membaca file
df_train = pd.read_csv(train_path)
df_test = pd.read_csv(test_path)

# Cek isi file
print("Train Data:")
print(df_train.head())

print("\nTest Data:")
print(df_test.head())
```

```python
 # 2. Hapus baris pada train yang memiliki nilai kosong pada kolom target

target_columns = ['task_achievement', 'coherence_and_cohesion', 'lexical_resource', 'grammatical_range']
df_train = df_train.dropna(subset=target_columns)
```

```python
# 3. Gabungkan kolom 'prompt' dan 'essay' menjadi satu kolom teks

df_train['text'] = df_train['prompt'] + " " + df_train['essay']
df_test['text'] = df_test['prompt'] + " " + df_test['essay']
```

```python
# 4. TF-IDF vectorization

tfidf = TfidfVectorizer(
    max_features=5000,   # ambil 5000 kata paling informatif
    stop_words='english' # hilangkan stopwords
)
X_train = tfidf.fit_transform(df_train['text'])  # hasil dalam bentuk sparse matrix
X_test = tfidf.transform(df_test['text'])        # transform test dengan vektorizer yang sama
```

```python
# 5. Ekstrak target (label)

y_train = df_train[target_columns]
```

```python
# 6. Cetak bentuk akhir

print("Shape X_train:", X_train.shape)
print("Shape X_test:", X_test.shape)
print("Shape y_train:", y_train.shape)
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.linear_model import Ridge
from sklearn.multioutput import MultiOutputRegressor
from sklearn.metrics import mean_squared_error
import numpy as np
import pandas as pd

# 2. Mount dan load data (sesuai kode kamu)
from google.colab import drive
drive.mount('/content/drive')

train_path = '/content/drive/My Drive/KELAS AI/df_train.csv'
test_path = '/content/drive/My Drive/KELAS AI/df_test.csv'

df_train = pd.read_csv(train_path)
df_test = pd.read_csv(test_path)

# 3. Drop NA pada target
target_columns = ['task_achievement', 'coherence_and_cohesion', 'lexical_resource', 'grammatical_range']
df_train = df_train.dropna(subset=target_columns)

# 4. Gabungkan prompt + essay
df_train['text'] = df_train['prompt'] + " " + df_train['essay']
df_test['text'] = df_test['prompt'] + " " + df_test['essay']

# 5. Load SBERT model
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')  # ringan dan akurat

# 6. Embedding teks menjadi vektor
X_train = model.encode(df_train['text'].tolist(), show_progress_bar=True)
X_test = model.encode(df_test['text'].tolist(), show_progress_bar=True)

# 7. Ambil target
y_train = df_train[target_columns]

# 8. Regressor (pakai Ridge, bisa diganti ke MLP/XGBoost)
regressor = MultiOutputRegressor(Ridge(alpha=1.0))
regressor.fit(X_train, y_train)

# 9. Prediksi dan evaluasi (opsional)
y_pred = regressor.predict(X_train)
rmse = np.sqrt(mean_squared_error(y_train, y_pred))
print(f"RMSE (Train): {rmse:.4f}")

# 10. Prediksi data test
y_test_pred = regressor.predict(X_test)
df_pred = pd.DataFrame(y_test_pred, columns=target_columns)
print(df_pred.head())
```
